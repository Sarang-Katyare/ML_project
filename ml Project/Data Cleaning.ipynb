## Splitting the data into train and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=20010)
```

## Train the model

```python
def train_model_lr():
    from sklearn.linear_model import LinearRegression
    model=LinearRegression()
    model.fit(x_train,y_train)
    return model
```

```python
def train_model_xg():
    from xgboost import XGBRegressor
    model=XGBRegressor()
    model.fit(x_train,y_train)
    return model
```

```python
def train_model_cb():
    from catboost import CatBoostRegressor
    model=CatBoostRegressor()
    model.fit(x_train,y_train)
    return model
```

```python
def train_model_nb():
    from sklearn.naive_bayes import GaussianNB
    model=GaussianNB()
    model.fit(x_train,y_train)
    return model
```

```python
def train_model_rf():
    from sklearn.ensemble import RandomForestRegressor
    model=RandomForestRegressor()
    model.fit(x_train,y_train)
    return model
```

```python
def train_model_ab():
    from sklearn.ensemble import AdaBoostRegressor
    model=AdaBoostRegressor()
    model.fit(x_train,y_train)
    return model
```

## Evaluate the model

```python
def evaluate_model(model,model_name):
    y_true=y_test
    y_pred=model.predict(x_test)
    from sklearn.metrics import r2_score
    r2=r2_score(y_true,y_pred)

    return model_name,r2
```

```python
models=[
    (train_model_lr(),'Linear Regression'),
    (train_model_xg(),'XG Boost'),
    (train_model_cb(),'Cat Boost'),
    (train_model_nb(),'Naive Bayes'),
    (train_model_rf(),'Random Forest'),
    (train_model_ab(),'AdaBoost')
]
```

```python
result=[]
for (model,model_name) in models:
    result.append(evaluate_model(model,model_name))  
```

```python
result_df=pd.DataFrame(result,columns=["Algorithm","R2 Score"])
result_df
```
