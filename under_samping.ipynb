```python
def drop_state_row(df, state, val):
    category_to_drop = state
    category_rows = df[df['state'] == category_to_drop]
    
    num_retain = int(len(category_rows) * val)
    
    retained_rows = category_rows.sample(n=num_retain, random_state=42)

    result_df = pd.concat([retained_rows, df[df['state'] != category_to_drop]])
    
    result_df = result_df.reset_index(drop=True)

    return result_df
```

```python
df1 = drop_state_row(df, "Victoria", 0.12)
```

```python
df2 = drop_state_row(df1, "New South Wales", 0.08)
```

```python
df3 = drop_state_row(df2, ' Western Australia', 0.2)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['state'] = encoder.fit_transform(df['state'])
```

```python
from sklearn.model_selection import train_test_split

x = df[['bedrooms', 'bathrooms', 'parking', 'area', 'state']]
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2863)
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, model.predict(x_test))
```

```python
from xgboost import XGBRegressor

model2 = XGBRegressor()
model2.fit(x_train, y_train)
```
